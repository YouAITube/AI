Let's discuss what gradient is and why descent is necessary.

The gradient of a function $f$ is an $n$-dimensional vector of partial derivatives.

$$ \nabla f(x_1, \ldots, x_d) = \left( \frac{\partial f}{\partial x_i} \right)_{i=1}^d. $$

For example, if a function depends on three variables: $F(x,y,z)$, then its gradient will be

$$ \nabla F(x,y,z) = \left( \frac{\partial F}{\partial x}, \frac{\partial F}{\partial y}, \frac{\partial F}{\partial z} \right). $$

Moreover, the gradient determines the direction of the steepest ascent of the function. Consequently, the anti-gradient will indicate the direction of its fastest decrease, which will be useful to us in our task of minimizing the error functional.

Gradient descent is a method of finding a local minimum by moving along the gradient.


```python
def func(x):
    return x ** 2

# grad_function
def gr_func(x):
    return 2 * x
```

Let's implement two functions:

mserror: Mean Squared Error function defined as $MSE = \frac{1}{n}\sum_{i=0}^n{(\text{y}_i-\text{y_pred}i})^2 = \frac{1}{n}\sum{i=0}^n{(\text{y}i-(w_1\cdot X_i + w_0)})^2 = \frac{1}{n}\sum{i=0}^n{(\text{y}_i-w_1\cdot X_i - w_0})^2$

gr_mserror: Gradient of the MSE function. We'll express it separately for the intercept coefficient and the slope coefficient:

Intercept:
$\frac{ MSE}{ w_0} = \frac{1 \cdot 2}{n}\sum({y_i -\text{y_pred}_i})\cdot -1$

Slope:
$\frac{ MSE}{ w_1} = \frac{1 \cdot 2}{n}\sum({y_i -\text{y_pred}_i})\cdot -X$

```python
def mserror(X, w1, w0, y):
    y_pred = w1 * X[:, 0] + w0
    return np.sum((y - y_pred) ** 2) / len(y_pred)

# grad_function
def gr_mserror(X, w1, w0, y):
    y_pred = w1 * X[:, 0] + w0
    return np.array([2/len(X)*np.sum((y - y_pred)) * (-1),
                     2/len(X)*np.sum((y - y_pred) * (-X[:, 0]))])
```

```python
# Set the minimum value by which weights should change
eps = 0.0001

# Initial points for w1 and w0
w1 = 0
w0 = 0

# Learning rate
learning_rate = 0.1

next_w1 = w1
next_w0 = w0

# Number of iterations
n = 100
for i in range(n):
    cur_w1 = next_w1
    cur_w0 = next_w0

    # Move in the negative direction of the computed gradient
    next_w0 = cur_w0 - learning_rate * gr_mserror(X, cur_w1, cur_w0, y)[0]
    next_w1 = cur_w1 - learning_rate * gr_mserror(X, cur_w1, cur_w0, y)[1]

    # Stop when the required degree of accuracy is reached
    print(f"Iteration: {i}")
    print(f"Current point {cur_w1, cur_w0}| Next point {next_w1, next_w0}")
    print(f"MSE {mserror(X, cur_w1, cur_w0, y)}")
    print("--------------------------------------------------------")

    if (abs(cur_w1 - next_w1) <= eps) and (abs(cur_w0 - next_w0) <= eps):
        break
```

For the convenience of implementing gradient descent, let's transition from the element-wise summation form ($MSE = \frac{1}{n}\sum_{i=0}^n{(\text{y}_i-\text{y_pred}_i})^2$) to the matrix form.

The predictions
The predictions of the linear model can be calculated as the dot product of the feature matrix and the weight vector plus the intercept term.
$$y_{pred} = X\cdot w + w_0$$

It's crucial to ensure that the dimensions of the matrix $X$ and the vector $w$ are compatible. In our case, they are the same.

```python
X.shape, coeffs.shape
```

To ensure that objects can be multiplied in matrix form, the number of columns in the first matrix should be equal to the number of rows in the second (though in our case, it's not a matrix but a vector).

In our case, they match, so we can multiply them, resulting in nothing but the dot product - all feature values are multiplied by their corresponding weights and then summed up.

So, we can rewrite the formula as:
$$y_{pred} = \langle X, w\rangle + w_0$$

However, the intercept term complicates matters. One workaround is to add a dummy feature to the data, which is equal to 1 for each object.

```python
coeffs
```

```python
X = np.column_stack([X, np.ones((50))])
X
```

And now all predictions of the linear model will be equal to:
$$y_{pred} = \langle X, w\rangle$$
And our MSE error will transform to look as follows:


$$MSE = \frac{1}{n}\sum^{n}_{i=1}(y_{i} - \text{y_pred}_i)^{2} = \frac{1}{n}\sum^{n}_{i=1}(y_{i} -  \langle X_i, w\rangle)^{2} =  \frac{1}{n}||Y -  X w||^{2}$$

where the $L_{2}$ norm is used:

$$||Y -  X w|| = \sqrt{\sum_{i=1}^n{(y_i-X_iw)^2}} $$

$$MSE = \frac{1}{n}\sqrt{\sum_{i=1}^n{(y_i-X_iw)^2}} ^{2} = \frac{1}{n}\sum_{i=1}^n{(y_i-X_iw)^2}$$



Let's implement two functions, but now using matrix operations:

mserror_mat: Mean squared error function for matrices.

gr_mserror_mat: Gradient function of the MSE for matrices:

$\frac{ MSE}{ w} = \frac{1 \cdot 2}{n}({Y - Xw}) \cdot-X$

```python
# Function to determine the mean squared error
def mserror_mat(X, w, y):
    y_pred = X @ w
    return np.sum((y - y_pred) ** 2) / len(y_pred)

# Gradient function
def gr_mserror_mat(X, w, y):
    y_pred = X @ w
    return 2/len(X)*(y - y_pred) @ (-X)
```

```python
#Initial weight
weights = np.zeros(X.shape[1])
weights
```

```python
# Set the minimum value by which weights should change
eps = 0.0001

# Learning rate
learning_rate = 0.1

next_weights = weights

# Number of iterations
n = 100
for i in range(n):
    cur_weights = next_weights

    # Move in the negative direction of the computed gradient
    next_weights = cur_weights - learning_rate * gr_mserror_mat(X, cur_weights, y)

    # Stop when the required degree of accuracy is reached
    print(f"Iteration: {i}")
    print(f"Current point {cur_weights}| Next point {next_weights}")
    print(f"MSE {mserror_mat(X, cur_weights, y)}")
    print("--------------------------------------------------------")

    if np.linalg.norm(cur_weights - next_weights, ord=2) <= eps:
        break
```

```python
print('Weights for features')
print(f'True {coeffs},\nTrained {next_weights[:-1]}')

print('\nIntercept weight', end=' ')
print(f'True 0, trained {next_weights[-1]}')
```
